```python
means_by_neighbourhood = df_Toronto.groupby('area').mean().reset_index()
medians_by_neighbourhood = df_Toronto.groupby('area').median().reset_index()
means_by_neighbourhood
```

```python
medians_by_neighbourhood
```

```python
# neighbourhood statistics from StatsCan
stats = pd.io.excel.read_excel('D:/Dropbox/SCS_3250/neighbourhoods.xlsx')
stats
```

```python
# Join average askprice, average soldprice, median askprice, median soldprice to neighbourhood statistics
median_df = stats.merge(medians_by_neighbourhood, left_on='neighbourhood',right_on='area', how='left').rename(columns={'askprice': 'median_askprice', 'soldprice': 'median_soldprice'})
median_df.drop(['city_number','area','type_number','bdrm','wshrm','latitude','longitude','maint'], axis=1, inplace=True)
final_df = median_df.merge(means_by_neighbourhood, left_on='neighbourhood',right_on='area', how='left').rename(columns={'askprice': 'mean_askprice', 'soldprice': 'mean_soldprice'})
final_df.drop(['city_number','area','type_number','bdrm','wshrm','latitude','longitude','maint'], axis=1, inplace=True)
final_df
```

```python
final_df['mean_soldprice/income'] = final_df['mean_soldprice']/final_df['mean_aftertax']
final_df['median_soldprice/income'] = final_df['median_soldprice']/final_df['median_aftertax']
```

```python
final_df.sort_values('mean_soldprice/income', ascending=False)
```

```python
x = final_df['mean_aftertax'][:]
y = final_df['mean_soldprice'][:]
fig, ax = plt.subplots()
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
ax.set_title('Toronto Neighbourhoods by Mean Income vs Mean House Sold Price',fontsize = 20)
ax.set_xlabel('Mean Income',fontsize = 20)
ax.set_ylabel('Mean Sold Price',fontsize = 20)

ax.annotate('y = ' + str(fit[0]) + 'x + ' + str(fit[1]), xy=(300000, 3500000), xytext=(200000, 3500000),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        if point['y'] > 700000:
            ax.text(point['x'], point['y'], str(point['val']))

label_point(x, y, final_df['neighbourhood'][:], ax)
```

```python
x = final_df['median_aftertax'][:]
y = final_df['median_soldprice'][:]
fig, ax = plt.subplots()
fit = np.polyfit(x, y, deg=1)
ax.plot(x, fit[0] * x + fit[1], color='red')
ax.scatter(x, y)
ax.set_title('Toronto Neighbourhoods by Median Income vs Median House Sold Price',fontsize = 20)
ax.set_xlabel('Median Income',fontsize = 20)
ax.set_ylabel('Median Sold Price',fontsize = 20)

ax.annotate('y = ' + str(fit[0]) + 'x + ' + str(fit[1]), xy=(155000, 2500000), xytext=(100000, 2500000),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        if point['y'] > 700000:
            ax.text(point['x'], point['y'], str(point['val']))

label_point(x, y, final_df['neighbourhood'][:], ax)
```
